## Disciplined process for building a data-driven linear differential equation model of a dynamic system using a step or initial condition response.

* Perform a step or initial condition reponse of the system. 
* Plot your data as a time series. Determine whether the system is stable for the output you measured, given the input you provided to the system (if any).
* If the system is stable, use a second test to determine whether a *linear* model is appropriate.
* If the system is stable, Determine what order model may be appropriate to describe your system's behavior. 
    - If the system does not show any significant transient dynamics on its way to steady state (it gets there immediately), your model may not need to be dynamic at all. Be very careful making this determination... no physical systems can change their energy state immediately. Only use a "$0^{th}$" order model if you are **certain** that the transient response of the system is too fast to worry about.
    - If the system shows exponentially decaying growth or exponential decay on its way to steady state, a first order model may be appropriate.
    - If the system shows exponentially decaying growth or exponential decay with an inflection point near $t=0$, a second order model may be appropriate.
    - If the system shows oscillatory behavior around a steady state, a second order model may be appropriate
    - If the system shows a combination of oscillatory and exponential behavior, a higher-order model may be appropriate.
* Once an appropriate model order has been selected, use a general form for the model's differential equation to determine the relationship between the data and the model's steady state value and/or gain. An automated fitting approach such as a least-squares curve fit can be used, but manual methods are also possible in many cases:
    - For first order models, you can use [This approach](../W03A_Modeling_MathematicalTools/W03A_Modeling_MathematicalTools.ipynb#Steady-State-Gain)
    - For second order models you can use [This approach](../W03A_Modeling_MathematicalTools/W03A_Modeling_MathematicalTools.ipynb#Step-Response-with-Zero-Initial-Conditions:-Repeated-Roots)
* Once an appropriate model order has been selected, use a general form for the model's differential equation to determine the relationship between the data and the model's parameters. An automated fitting approach such as a least-squares curve fit can be used, but manual methods are also possible in many cases:
    - For first order models you can use [This approach](../W03A_Modeling_MathematicalTools/W03A_Modeling_MathematicalTools.ipynb#Time-Constant).
    - For underdamped second order models you can use [This approach](../W03A_Modeling_MathematicalTools/W03A_Modeling_MathematicalTools.ipynb#Characterizing-Second-Order-Responses).
* Once the model is fully characterized, use a standard form for the solution to compare your model with the your data.
* Validate your model by comparing a *second* test with a *different* input to your model *without* fitting new parameters.

## Using a Differential Equation Model for Controller Design

As you may recall from [Week02B](../W02B_Modeling_Fundamentals/W02B_Modeling_Fundamentals.ipynb), one of the original goals of this challenge was to develop a system to *regulate* the voltage at the house to make sure that the charging station has the proper voltage. 

As you may recall from ME352, a generator's output voltage is directly proportional to its speed. Because the dam's generator is directly connected to its turbine shaft in our simulation, we can cast the problem of regulating the house voltage as a problem of regulating *turbine speed*. We can regulate turbine speed by using a feedback controller.

The type of control we will focus on in ME480 is called *classical feedback control*. In classical feedback control, information about a system's *output* is "fed back" to the controller, which may be an Arduino or similar microprocessor, in order for it to make "informed decisions" about what to do next. A graphical representation of this paradigm is shown as a *block diagram* below.

In the diagram above, the controller implements a control *law* to turn the **error** between $\Omega_d$, or the "desired angular speed of the turbine," into *action* that it takes in order to influence the turbine. If the system could do this "on its own," it could automatically account for variations in the dam height due to rainfall or drought, keeping the turbine spinning at the correct speed all the time!

The most obvious action the dam system could take to influence the turbine's speed is to change the penstock valve position to control the *fluid resistance* of the valve. Closing the valve slightly could increase the resistance, slowing the turbine down by providing it less fluid power. Opening the valve slightly would decrease the valve's resistance, speeding the turbine up. 

By making "decisions" about how much to open or close the valve based on the *error* $e$ between desired and actual turbine speed, it is possible for the controller to regulate the turbine's speed automatically. This implies that the arrow representing the *input* to the dam system should be *valve position* $u_v$, as shown below.

Looking at this diagram, one can see that the overall input to the system is desired turbine speed, and the output of the overall system is actual turbine speed. With a good design for the controller, these two should match fairly well, no matter what happens. Looking at the individual "pieces" of this diagram, we can see that the "controller" should have an input of error and an output of valve position, and the "dam system" should have an input of valve position and an output of turbine speed.


This sounds great, but there is a problem. In order to *design* a good controller for the dam system, we need to have a *model* for the dam that we trust, and one that we can work with using linear differential equations tools.

Think back to your initial investigation of the relationship between valve position and house voltage from the second problem in the [Week02 assignment](../W02E_Exercises/W02E_Exercises.ipynb). There, you found *based on observation* that doubling the valve position with a particular dam input pressure did not double the system output. This means that **the dam system is nonlinear** for our choice of input and output.

In the following sections, we will discuss how we can deal with nonlinearities in differential equation models for dynamic systems.

## Recognizing and Handling Nonlinearities in System Differential Equations

Differential equations are linear if they satisfy the principle of [superposition](../W02B_Modeling_Fundamentals/W02B_Modeling_Fundamentals.ipynb#Linearity:-the-principle-of-superposition). All linear, constant-coefficient differential equations can be written in the compact form:

$$\sum_{n=0}^{n=N} a_n \frac{d^n y}{dt^n} = \sum_{m=0}^{m=M} b_m \frac{d^m u}{dt^m} $$

Where $M<=N$, $u(t)$ is the equation's input, and $y(t)$ is the equation's output. We call this an "input-output" differential equation. This is slightly different than the compact form we saw [here](../W03A_Modeling_MathematicalTools/Week03A_Modeling_MathematicalTools.ipynb#Solving-Differential-Equations:-the-method-of-undetermined-coefficients) but this one allows us to consider a known input whose derivatives might also be important for describing the system's behavior. Theoretically, if the input function is known, its derivatives are also known, and the "total input" to the system could be redefined so that no input derivatives exist. However, we will consider the more general case here, and use the form above.

**For this discussion**, we will assume that you already have your differential equation model in input-output form. 

We can recognize nonlinearities in an input-output differential equation by recognizing **any deviation** from the form above. Some common nonlinearities include:
* A constant term (one that is not multiplied by an output, output derivative, input, or input derivative). 
* A term that is a nonlinear function of the input, a derivative of the output, or a derivative of the output (e.g. $\dot{y}^2$ or $\left| y \right|$). 
* Single term that includes both the input and the output (e.g. $u\cdot y$) 

## Handling Nonlinear Behavior

To handle nonlinearity in your model, you have two basic choices:

1. Solve (or simulate) the nonlinear differential equation and use the nonlinear model as-is. 
2. Linearize the model about a normal operating point. For data-driven models, this essentially means "use a linear fit for the system's behavior if possible." For physics-based, parametric models, linearization can be performed using the Taylor Series Expansion.

Some types of nonlinearities are very easy to handle using linear differential equations techniques. For example, "extra" constant terms can sometimes be treated as separate inputs, and then the "total" solution to the differential equation is the sum of the particular solutions to each. By re-imagining these terms as inputs, the system can be treated as a linear system by using superposition to find several particular solutions, and adding them together.

On the other hand, some types of nonlinearities cannot be handled using linear differential equations techniques at all. For example, the term $\left|y\right|$ has a discontinuity at $y=0$. The only option is to "solve the nonlinear differential equation (option 1)." In ME480, our only option for this type of situation is to arrive at a solution using **numerical integration**, which is covered below.

With any luck, however your system operates in response to a *small range* of inputs that are all in the *neighborhood* of some average known input, **is stable** for the range of inputs your system will experience, and thus the system's output remains in a small *neighborhood* of some average known output. We call the combination of an average known input and an average known output the **Normal Operating Point** of your system. You may be able to fit a linear differential equation to the system's behavior as long as the deviation of your input $\Delta u$ from an average input value $\bar{u}$ is small.

## Linearizing Nonlinear Models

Once you know you have a nonlinear system on your hands whose nonlinearity may be significant, you have two options for creating a useful linear model that you can try.

1. If you have a nonlinear differential equation (possibly from a physics-based modeling process) then your model can be linearized using the [Taylor Series Expansion](https://courses.engr.illinois.edu/ece486/fa2017/documents/lecture_notes/state_space_p2.pdf). We will not spend much time on this in ME480, but a conceptual understanding of this process is important, even when we are only working with models derived from data. 

2. If all you have is data to begin with, and you did not start with a physics-based model, then you might fit a linear differential equation model to your data, even if you suspect that the system is nonlinear. To do this well, you make sure that you only expect your model to work for **small changes** in the input, and therefore only describe the resulting **small changes in the output** that occur. Your model should subtract any initial conditions and/or inputs, and only model change in input $\rightarrow$ change in output.

The idea in both approaches to "linearization" is the same. The concept is that we create a *linear model* that approximates the nonlinear model for a small range of behavior around a "normal operating point," or NOP. We can write this formally by saying that any *total* input to the system $u$ can be written as:

$$u = \bar{u} + \Delta u$$

where $\bar{u}$ is an average, constant input, and $\Delta u$ is some small perturbation around this input. This results in some system behavior $y$ that we can write as:

$$y = \bar{y} + \Delta y\text{ ,}$$

Where $\bar{y}$ is the system's "prior steady state," a constant initial condition that occurred in response to $\bar{u}$ before $\Delta u$ was applied, and $\Delta y$ is the small change in behavior that results from $\Delta u$. When we linearize a first-order model this way, for example, we go from an original differential equation relating input $u$ to output $y$:

$$\frac{d}{dt} y = f_{NL}(y,u)$$

where $f_{NL}$ might be a nonlinear or partially nonlinear function, to a model that is linear but only applies to the relationship between *change in input* $\Delta u$ $\rightarrow$ *change in output* $\Delta y$:

$$\frac{d}{dt} \Delta y = f(\Delta y, \Delta u)$$

Where $f$ is a linear function. This linearized model is a model for *incremental* variables representing small changes, $\Delta u$ and $\Delta y$, **not** a model for total input $u$ or total input $y$. Think of this like a "coordinate shift" for the system.

### Example: Difference between incremental behavior around an NOP and total behavior in dynamic system test data

Consider an example where you wanted to model a car's forward velocity $v$ in response to its throttle input percentage, $T_p$. Let's say you collected data for the car's velocity. These data for the input ($T_p$ in %) and the output ($v$) are shown in the plot below.

````octave
clear all
% load the dataset. time, Tp, v are the columns
data_dT10 = load('cardata_deltaT_10.txt');
% pull out the variables we're interested in.
t10 = data_dT10(:,1);
Tp10 = data_dT10(:,2);
v10 = data_dT10(:,3);
% plot data
figure
subplot(2,1,1)
plot(t10,Tp10,'k-o')
xlabel('Time (s)')
ylabel('Throttle % T_p','interpreter','TeX')
subplot(2,1,2)
plot(t10,v10,'k')
xlabel('Time (s)')
ylabel('vehicle speed v (m/s)','interpreter','TeX')
````

At time $t=0$ seconds, the throttle was abruptly changed from 0% to 50% while the car was at rest. Then, the car was allowed to reach a steady state speed. At $t=10$ seconds, the throttle was abruptly changed from $50\%$ to $60\%$. We could say that the second change in throttle represented a **change** $\Delta T_p$ over an average, initial value $\bar{T_p}$. If we suspected that the car exhibited nonlinear behavior, we might want to create a linearized model that only describes the car's behavior around the normal operating point $T_p=50$. 

Now, instead of looking at the true vehicle speed and throttle position, we could choose to model only the *incremental* behavior of the system beyond its initial steady state throttle input and speed. We can accomplish this by subtracting the step time and initial offsets from our data, effectively creating a dataset that describes variables $\Delta T_p$ and $\Delta v$.

````octave
% use original dataset to determine where to chop
steptime = 10;%second step input occurs at t=10 seconds
stepindex = find(t10>=steptime,1,'first');%find the index (row in file) at which the step occurs 

% find Tpbar and vbar, the NOP. The NOP is the steady state that occurs right before the new step... at our "new" t=0.
Tpbar10 = Tp10(stepindex)
vbar10 = v10(stepindex)

% convert data to incremental variables by subtracting Tbar and vbar (our NOP)
dt10 = t10(stepindex:end)-t10(stepindex);%create a new time vector starting at zero that only looks at the incremental second step input.
dv10 = v10(stepindex:end)-vbar10;%chop the data and subtract the initial condition (the former steady state or NOP) vbar
dTp10 = Tp10(stepindex:end)-Tpbar10;%chop the data and subtract the initial condition (the former steady state or NOP) Tpbar

%plot the test data in terms of incremental variables
subplot(2,1,1)
plot(dt10,dTp10,'k-o')
xlabel('Time (s)')
ylabel('Throttle change % \Delta T_p','interpreter','TeX')
subplot(2,1,2)
plot(dt10,dv10,'k')
xlabel('Time (s)')
ylabel('vehicle speed change  \Delta v (m/s)','interpreter','TeX')
````

Here, we see that we are now only considering the *change* in speed with respect to a *change* in steady throttle input. Pre-processing our test data this way will help us if our system is nonlinear, because we may be able to use a linear model to describe our system's behavior as long as we stick to trying to model *small changes* in system input and output.

### Linearizing nonlinear models using a parametric equation and the Taylor Series
Let's imagine that you used techniques you learned in ME352 to develop a nonlinear differential equation model for a car's velocity. This model might take an input of throttle %, $T_p$, and result in a forward velocity for the car, $v$. With parameter values substituted in, maybe this car's model turns out to be:

$$\frac{d}{dt} v = -p_1 v \mathop{\underset{\mmlToken{mo}{}}{- p_2 v^2}}\limits_{F_{NL}(v)} + p_3 T_p$$

Where $p_1,p_2$ and $p_3$ are physical parameters that include energy storage, transduction, and dissipation terms in the car (maybe they're made up of ratios of damping constants to mass terms, for example).

In looking at this model, you consult [this section](#Recognizing-and-Handling-Nonlinearities-in-System-Differential-Equations) of the notebook, and you realize that this system is *nonlinear* because it includes a nonlinear function of the system's output, vehicle speed in the term $F_{NL}=-p_2v^2$. **Note that many other types of nonlinearities are possible. This is just an example.** 

In this case, we see that doubling the throttle input $T_p$ on the vehicle *will not* result in double the final speed, because of this nonlinearity. If you were to plot the nonlinear term $F_{NL} = -p_2v^2$ vs. $v$ on a plot, it might look like this:

The red line represents the nonlinear term, and the blue line represents a "linear approximation" for some average (constant) input $\bar{T}_p$ and some resulting average speed $\bar{v}$. Using the first term of the [Taylor Series](https://courses.engr.illinois.edu/ece486/fa2017/documents/lecture_notes/state_space_p2.pdf) expansion, we can compute the derivative of the nonlinear function and come up with the equation for a line that is a "pretty good" approximation of the nonlinear function for small perturbations $\Delta T_p$ around the average throttle input $\bar{T}_p$, which gives a *total* throttle input to the car of $T_p = \bar{T}_p+\Delta T_p$.

Following the procedure for linearization of a nonlinear model using the Taylor Series, you might find that the original, nonlinear model can be *linearized* to become a model relating $\Delta T_p$, or the small change in throttle input about a constant, average value, to $\Delta v$, or the small change in vehicle velocity around an average constant "initial" velocity $\bar{v}$. This nonlinear model might look like:

$$\frac{d}{dt} \Delta v = -p_1\Delta v - p_{2a}\Delta v + p_3 \Delta T_p$$

Where the new constant *linear* parameter $p_{2a} = 2p_2\bar{v}$. The model is now linear, but it only describes *changes* in vehicle velocity in response to *changes* in throttle input about the initial, constant conditions.

To see how this linear model compares to the data we collected, we can use our original dataset and the following estimates for the model parameters, whose units will ensure that all terms in the equation have units of $\frac{m}{s^2}$ when multiplied through, since those are the units of the left-hand side of the equation, which represents the car's change in acceleration.

* $p_1 = .1 \left(\frac{1}{s}\right)$
* $p_2 = .02 \left(\frac{1}{s}\right)$
* $p_3 = 0.25 \left(\frac{1}{s^2Nm}\right)$;

Using a [standard form](../W03A_Modeling_MathematicalTools/W03A_Modeling_MathematicalTools.ipynb#First-Order-Linear-Differential-Equations) for a first-order differential equation:

$$a_1\dot{y} + a_0y= u(t)$$

We can see with a little bit of simple algebra that for our differential equation, $a_1=\frac{1}{p_3}$, and $a_0=\frac{p_1+p_{2a}}{p_3}$.

Now using our standard form for the solution to a zero-initial-condition [first-order step response](../W03A_Modeling_MathematicalTools/W03A_Modeling_MathematicalTools.ipynb#Step-Response), we can plot our final, linearized model against the data. This is shown in the cell below.

````octave
% note that the data were loaded in an earlier cell and converted to incremental variables. NOP was also calculated.

p1 = .1;
p2 = .02;
p3 = .25;
% now compute p2a which we got from our linearization
p2a = 2*p2*vbar10;

% now compute a0 and a1 for use in the standard form:
a1 = 1/p3;
a0 = (p1+p2a)/p3;
U = dTp10(end)%this gives me the value for the input based on the data I collected. I'll just use the last value since it was a constant.

% now, we can run our model against the simulation.
delta_t = .001;
tsim = 0:delta_t:dt10(end);%create a simulation time vector that's as long as the dataset.

%compute the solution to our linearized differential equation.
dvsim_linearized = U/a0*(1-exp(-a0/a1*tsim));

%now compare apples to apples. Compare data for delta_v vs. dvsim_linearized.
figure
plot(dt10,dv10,'k',tsim,dvsim_linearized,'r')
xlabel('Time (s)')
ylabel('\Delta v (m/s)')
legend('Data','Linearized Model Simulation')
````

As you can see, the linearized model fits the data well for this small 10% increase in throttle percentage, which both model and data show result in a steady state change in car velocity of around 2.5 m/s or around 5mph. 

### Creating a Linearized Model from Data Alone

If you had no parametric (e.g. physics-based) model to begin with, you may be able to create a linear model from a dataset directly. This is what we will be doing most commonly in ME480. The procedure is to:

* Collect a step response dataset about some NOP using a small change in input.
* [Convert the dataset to incremental variables](#Example:-Difference-between-incremental-behavior-around-an-NOP-and-total-behavior-in-dynamic-system-test-data) (subtract initial conditions) so that you are only looking at the output that results from a relatively small change in input, where the incremental dataset begins with zero initial conditions.
* Fit a linear differential equation model to the dataset using the [disciplined process](#Disciplined-process-for-building-a-data-driven-linear-differential-equation-model-of-a-dynamic-system-using-a-step-or-initial-condition-response.) above.
* Validate your model using a second dataset where data were collected using a different step size. Try to use the maximum step size that you expect your system to be subjected to.

We can apply this procedure to the car example as shown below. Note that we will be using variables defined in earlier cells to highlight the *new* parts of this procedure.

## *Model Validation* for Linearized Models

In order to determine whether your linearized model is an appropriate approximation of the nonlinear model, you can do one or both of the following, as appropriate:

1. Compare the response of your linearized model to the response of the nonlinear model using *numerical integration* to simulate the response of the nonlinear model. This only applies if you had a parametric (e.g. physics-based) model that was linearized using the Taylor Series Expansion. This is what we call an *internal validity* check, to make sure the linearization procedure was performed correctly. If at all possible, you should **also** complete step 2, which tests *external validity*.
2. Compare the response of your linearized model to the *real system's* response by performing an experiment, and comparing data to the response of your linearized model. You **should never** use the same dataset for both fitting and validation. Validation is about seeing whether your model *generalizes* to some new scenario. A model that generalizes to new circumstances is said to show *external validity*. Collect a second dataset for validation using a different input magnitude, and compare this dataset with your linearized or data-based linear model.

As you work through validation, you must take care to remember that your *linearized* model is a model for the dynamic response of the *incremental output* $\Delta y = y - \overline{y}$ as a result of applying an *incremental input* $\Delta u = u - \overline{u}$. 

## Numerical Integration

Sometimes, a system does not lend itself to a linearized model well. This is often because the system has a discontinuous (non-differentiable) nonlinearity. In cases like these, the best option may be to use **numerical integration** to solve a nonlinear differential equation model for your system.

 Although numerical integration is **the only tool we'll use for systems that are nonlinear**, it is also perfectly fine to simulate a **linear** differential equation using numerical integration rather than using an analytical solution. 

To begin, we will review what we mean by "Integral," and then we will introduce Euler's Method for numerical integration.

### Integrals: Definition

For our purposes in this course, we will define the integral of a function $y(t)$, using the notation $\int y(t) dt$, as follows:

The function $Y(t)$ is the *integral* of a function $y(t)$ if $\frac{d}{dt} Y(t) = y(t)$ for all values of $t$.

This means, in simple terms, that the derivative of $Y(t)$ is $y(t)$. Integrals are often visualized or introduced in calculus classes as "the area under a curve," which is certainly a useful way to conceptualize them. Consider, for example, the scenario in the graphic below, which explains the relationship between a constant and its integral.

When the functional form of $y$ is known, it is possible to compute an *indefinite integral* for it, yielding the function $Y$ in our definition. However, for an arbitrary dataset, whose functional form may not be known, often the best we can do is to *approximate* the integral of a function. 

### Approximate Integration: Euler's Method

In Euler's method of approximate integration, also called the "rectangular approximation" of an integral, The area under a curve, $y(t)$, which yields $\int y(t) dt = Y(t)$, can be *approximated* by repeatedly summing the area of rectangles  projected either forwards or backwards in time from each discrete measurement of $y$. The process begins with some initial, known value of the integral $Y_0$, the value of $Y(t)$ is incremented by the area of the corresponding rectangle on the plot of $y(t)$. The process is shown visually below.

As you can see in the second plot, the idea is to use one, known value of our integral, $Y$, at the very start of the procedure. This is often called the *initial value* of $Y$. Then, the area under the derivative plot from $y(t_1)$ to $y(t_2)$ is added to our estimate of $Y$, and this value, when added to $Y_0$, becomes $Y(t_2)$. Formally, this can be written as:

$Y(t_2) = Y_0 + y(t_1)\cdot(t_2-t_1)$

**Aside: Compare this to what you know about point-slope form for a line: $y = m(x-x_0) + y_0$, with $x_0$ and $y_0$ describing a known point on the $y$ (dependent variable) vs. $x$ (independent variable) plot. It's the same equation!**

From $t_2$ to $t_3$, the procedure is repeated. In other words, the rectangle defined by the value of $y(t_2)$ is used to add "a little more" to our estimate to produce $Y(t_3)$. Formally,

$Y(t_3) = Y_0 + y(t_1)\cdot(t_2-t_1) + y(t_2)\cdot(t_3-t_2)$

Using a "recursive formulation," we could equivalently write:

$Y(t_3) = Y(t_2) + y(t_2)\cdot(t_3-t_2)$

And so on. **In this description, we projected the value of $y$ forwards in the creation of our approximate rectangles. Therefore, this particular variant of Euler integration is called "Forward Euler" integration.** "Backward Euler" integration is also possible. As with approximations of derivatives, the smaller the differences between times we use in the projection of the rectangles, the better our approximation of the derivative will be.

Because approximate integration is often performed on data arranged as arrays, we can use the counting index $k$ to write this as an equation that could be built into a "for loop" and computed efficiently in code. Considering arrays representing time $t$, derivative $y$, and integral $Y$, we could write an equation for any entry in the output array $Y(k)$ as:

$Y(k) = Y(k-1) + y(k-1)\cdot(t(k)-t(k-1))$

Note that this equation is an equation for a line in point-slope form. We use each estimate for $Y(k-1)$ to produce an estimate for the *next* value of our integral $Y(k)$ using linear extrapolation. This is shown visually on the right panel of the plot above.

An example of how to implement approximate Euler integration in MATLAB/Octave is shown below, given a vector of known values for $y$. The approximate integral is compared with the "true" integral of the function to show its accuracy. Try changing the timestep $T$ in the code cell below to see how accuracy is affected by the size of the time step used for extrapolation/integration. A large value of $T$ is problematic because it means you are extrapolating too far, sacrificing accuracy!

````octave
T = 0.01; %timestep
t = 0:T:1; %time vector
y = cos(10*t); %function we wish to integrate. This may come from data rather than a known function of t.

Y_estimate = zeros(size(t)); %initialize an array to hold our estimates for Y, the integral of y.
Y0 = 0; %some initial, known value for the integral at the beginning.
Y_estimate(1) = Y0; %place this value in the first entry of our array

Y_true = 0.1*sin(10*t); %for this example, we know that d/dt(sin(t)) = cos(t). 
%We will use this "exact" solution to the integral to allow us to see how accurate Euler integration is.

%now set up a for-loop starting at the second entry in Y_estimate, since we already know Y_estimate(1)=Y0
for k=2:length(t) %automatically set up the for-loop to be the right length
    %now implement Euler's formula. remember that here, y is dY/dt.
    Y_estimate(k) = Y_estimate(k-1) + y(k-1)*(t(k)-t(k-1));
end

%now we will compare our approximate integral with our "true" value to see if Euler integration works.

figure
plot(t,Y_true,'k.',t,Y_estimate,'r')
xlabel('Time (s)')
ylabel('Y(t)')
legend('true','Forward Euler Integration')
````

## Euler's method for differential equations

### Introduction

One way to deal with nonlinear system models, especially if all we want to do is visually see how they react to inputs, or compare a model with data, is to integrate their equations of motion numerically. Because the system is nonlinear, we can't use the linear differential equations techniques to easily obtain a closed-form solution to the system's behavior. Therefore, Numerical integration using something like [Euler Integration](https://en.wikipedia.org/wiki/Euler_method) or [Runge Kutta Integration](https://en.wikipedia.org/wiki/List_of_Runge%E2%80%93Kutta_methods) is usually a good option for investigating system response. 

Imagine that you produced a model $\frac{dy}{dt} = -4y^2 +2$. Because of the $y^2$ term, it isn't easy to obtain an analytical solution $y(t)$ using the method of constant coefficients, since the equation is not linear (doubling the input $u$ will not double the output $y$). 

###  Euler's method for differential equations: implementation

Because our output derivative $\frac{dy}{dt}$ depends on the current value of $y$ in a differential equation, we have to *use our most recent estimate for $y$* when computing our derivative $\frac{dy}{dt}$. This has to happen at every time step! See the example below, which uses Euler's method to solve our differential equation.

````octave
clear all
%create a time vector
t = 0:.001:1;
%create a vector of zeros to hold our y estimates
y = zeros(size(t));
%declare the first element in "y" as our best guess for the FIRST value of y (could be from a dataset, or knowledge of an experiment)
y(1) = 0.2;

%loop through time values to perform integration. start at the second value of time, since we already have a guess for y at the first value!
for k=2:length(t)
    %first use the differential equation to compute dy/dt (slope) based on our best guess for y at the previous time
    dydt = -4*y(k-1)^2 + 2; %2 is a constant, but it doesn't have to be. I could have a vector of known values for 2 as well.
    %now extrapolate (integrate) to our current time to get a new estimate for y!
    y(k) = dydt*(t(k)-t(k-1)) + y(k-1);
end

figure
plot(t,y,'k')
xlabel('Time (s)')
ylabel('y')
````

### Simulating Second Order Differential Equations in Input-Output Form Using Euler Integration

To simulate a second order differential equation using Euler integration, consider our introduction to Euler integration as extrapolation above, where computing the integral of a function was visualized as a process of repeated linear extrapolation over small timesteps.

If we consider a system model that is a *second order* input-output differential equation that takes the form:

$$\ddot{y} = f(\dot{y},y,u)$$

where $u$ is our system's input (source of power from outside of our scope), $y$ is our system's output (dependent variable), and "$f$" represents our **model** equation relating $\ddot{y}$ to $\dot{y}$ and $y$ and $u$, numerically integrating our function for $\ddot{y}$ could be used to produce an estimate of $\dot{y}$. Then, numerically integrating $\dot{y}$ one more time could be used to produce an estimate of $y$. 

This process of integrating $\ddot{y}$ once, then another time, to produce our final estimate of our model's output is attractive, and could be represented by the figure below, in which $\ddot{y}(k-1)$ is used to extrapolate an estimate for $\dot{y}$, and $\dot{y}(k-1)$ is used to extrapolate an estimate for $y(k)$.

The difficulty with this approach is that $\ddot{y}$ *depends on our last estimate of $\dot{y}$, $\dot{y}(k-1)$ as well as our last estimate of $y$, $y(k-1)$ and our last input $u(k-1)$.* This means that the two integration operations have to occur in the *same* loop. The structure of such a loop can be summarized as follows, where  `yd` represents $\dot{y}$ and `ydd` represents $\ddot{y}$:

```octave
for k=2:length(t)
    % compute second derivative of output based on model equation, represented here by "f."
    ydd(k-1) = f(yd(k-1),y(k-1),u(k-1));
    % now extrapolate once to get our new estimate of the first derivative of y
    yd(k) = yd(k-1) + (t(k)-t(k-1))*ydd(k-1);
    % now extrapolate a second time to get our new estimate of y itself.
    y(k) = y(k-1) + (t(k)-t(k-1))*yd(k-1);
end
```

Notice that we are are not using the *new* estimate of $\dot{y}$ in the second integration! See the figure above to see why-- we are calculating the slope of each function one timestep behind where we would like to predict *to*.

Also note that for this approach to work, an initial condition for $y$ is needed along with an initial condition for $\dot{y}$ so that the values `y(1)` and `yd(1)` are known, and can be used as the starting point for extrapolation on the first iteration of the loop.

Extending the integration methodology above to include systems with even higher order derivatives also works, as long as the model equation is solved for the *highest* derivative. Then, the loop structure presented above can be modified to perform as many sequential integrations as is necessary.

### Example: Integrating a second order differential equation using Euler's method

This approach is applied to our example in the code cell below. Let's say we constructed a physics-based model of an electrical circuit with an input of some voltage $V_1$ and an output of a different voltage $V_3$ that ended up being:

$$\ddot{V}_{3g} = \frac{1}{R_1C_1R_2C_2}(V_s - (R_1C_1+R_2C_2)\dot{V}_{3g} -V_{3g})$$

We could simulate this model using numerical integration using the methodology above.

````octave
clear all
%create a time vector
t =0:.1:20;

%set up independent variables
R1 = 1000;
R2 = 15000;
C1 = 470e-6;
C2 = 330e-6;
%input voltage
Vs = 5;
%initialize V3 and its derivative... both of which we need to keep track of.
V3d = zeros(size(t));
V3 = zeros(size(t));
for k=2:length(t)
    % compute second derivative of output based on model equation.
    V3dd = 1/(R1*C1*R2*C2)*(Vs - (R1*C1+R2*C2)*V3d(k-1)-V3(k-1));
    % now extrapolate once to get our new estimate of the first derivative of V3
    V3d(k) = V3d(k-1) + (t(k)-t(k-1))*V3dd;
    % now extrapolate a second time to get our new estimate of y itself.
    V3(k) = V3(k-1) + (t(k)-t(k-1))*V3d(k-1);
end

%plot results
figure
plot(t,V3,'k')
xlabel('Time (s)')
ylabel('Voltage V3')
````
